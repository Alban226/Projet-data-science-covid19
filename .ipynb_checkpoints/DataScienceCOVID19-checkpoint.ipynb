**Test d'hypothèse**
H0L :es individus atteints du covid 19 ont des taux moyens de leucocytes, plaquettes,monocytes et eosiphils égaux aux autres

```python
from scipy.stats import ttest_ind
```

```python
positive_df.shape
```

```python
negative_df
```

Pour utiliser le t_test il faut des classes équilibrés

```python
equi_ned = negative_df.sample(positive_df.shape[0])
```

```python
def t_test (col) :
    alpha = 0.02
    _,p = ttest_ind(equi_ned[col].dropna(),positive_df[col].dropna())
    if p < alpha :
        return 'H0 rejetée'
    else : 
        return 0
```

```python
for col in blood_columns:
    print(f"{col :-<50} {t_test(col)}")
```

Le t-test confirme bien le fait que les patients atteints du covid 19 n'ont pas en moyenne le même taux sanguins(monocytes et autres) que les autres patients significativement.

**Pre-processing**

```python
df = data.copy()
df.columns
```

**Supression des colonnes inutiles**

```python
df = df.drop(['Parainfluenza 2', 'Influenza B, rapid test', 'Influenza A, rapid test'], axis=1)
```

```python
df.columns
```

```python
df = df.drop(['Patient addmited to regular ward (1=yes, 0=no)',
       'Patient addmited to semi-intensive unit (1=yes, 0=no)',
       'Patient addmited to intensive care unit (1=yes, 0=no)'], axis=1)
```

**Creation d'un train_set et d'un test_set** 

```python
from sklearn.model_selection import train_test_split
```

```python
key_columns =['Patient age quantile', 'SARS-Cov-2 exam result','has_disease']
```

```python
key_columns = list(key_columns)
blood_columns = list(blood_columns)
df = df[key_columns + blood_columns]
```

```python
train_set,test_set = train_test_split(df, test_size=0.2, random_state = 0)
```

```python
train_set['SARS-Cov-2 exam result'].value_counts()
```

```python
test_set['SARS-Cov-2 exam result'].value_counts()
```

**Encodage des variables catégorielles**

```python
def encodage (df) :
    code = {'positive': 1, 'negative':0,
             'detected':1, 'not_detected':0}
    for col in df.select_dtypes('object') :
        df[col] = df[col].map(code)
    return df
```

**Imputation**

```python
def imputation(df) : 
    return df.dropna(axis=0)
    
```

**Fonction de preprocessing**

```python
def preprocessing(df) : 
    df = encodage(df)
    df = imputation(df)
    X = df.drop('SARS-Cov-2 exam result', axis=1)
    y = df['SARS-Cov-2 exam result']
    print(y.value_counts())
    return X, y
```

```python
X_train, y_train = preprocessing(train_set)
```

```python
X_test, y_test = preprocessing(test_set)
```

Pour obtenir des bons résultats nous allons utiliser le cycle suivant : 
idée->code->evaluation->idée

**Modelisation**

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.pipeline import make_pipeline
from sklearn.feature_selection import SelectKBest, f_classif
```

```python
model = make_pipeline(SelectKBest(f_classif, k=4),
                      RandomForestClassifier(random_state=0) )#Pour avoir le repetabilité
```

**Procedure d'évaluation**

```python
from sklearn.metrics import f1_score, classification_report, confusion_matrix
```

```python
from sklearn.model_selection import learning_curve
```

```python
def evaluation (model):
    model.fit(X_train, y_train)
    ypred = model.predict(X_test)
    print(confusion_matrix(y_test, ypred))
    print(classification_report(y_test, ypred))
    N, train_score, val_score = learning_curve(model, X_train, y_train, cv=4, train_sizes = np.linspace(0.2,1,5),scoring='f1')
    plt.figure(figsize=(12,8))
    plt.plot(N, train_score.mean(axis=1), label='train_score')
    plt.plot(N, val_score.mean(axis=1), label='val_score')
    plt.legend()
    plt.show()
    return
```

```python
evaluation(model)
```

```python
pd.DataFrame(model.feature_importances_, index=X_train.columns).plot.bar()
```

**1er Constat :** Les premières opérations basiques de preprocessing et de modelisation donne déja un model en overfitting
mais constate également que le val_score augment avec les données, ce qui veut dire qu'on a besoin de plus de données.
Aussi avec l'observation des features_importances, on constate que les variables virales n'ont pas d'effet sur le resulttat.
Je vais donc reentrainer le modèle sans les variables virales.

**2eme Constat :** Même en réduisant les variables le modèle est toujours en overfitting et deplus les performances du modèle DecisionTree ont baissé.


**RandomForestClassifier**

**1er Constat**: En gardant, toutes les variables le modèle est toujours en everfitting avec un recall faible pour les cas positifs.

**2eme Constat**: En retirant les variables virales, les résultats sont toujours aussi mauvais.
